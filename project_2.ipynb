```python
#import libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#load dataset
df = pd.read_csv("insta.csv")
print(df.head())
#basic info
#print(df.info())
#print(df.describe())
#mean and standard deviation
print("Mean Likes:", df['Likes'].mean())
print("Standard Deviation:", df['Likes'].std())
#correlation
correlation = df.corr(numeric_only=True)
print(correlation)
sns.heatmap(correlation, annot=True)
plt.title("Correlation Heatmap")
plt.show()
#Distribution of likes
sns.histplot(df['Likes'], kde=True)
plt.title("Likes Distribution")
plt.show()
#post type analysis
sns.barplot(x='Post_Type', y='Likes', data=df)
plt.title("Post Type vs Likes")
plt.show()
#hashtag vs likes
sns.scatterplot(x='Hashtags', y='Likes', data=df)
plt.title("Hashtags vs Likes")
plt.show()
#hypothenses testing
from scipy.stats import pearsonr
corr, p_value = pearsonr(df['Hashtags'], df['Likes'])
print("Correlation:", corr)
print("P-value:", p_value)
if p_value < 0.05:
    print("Reject Null Hypothesis: Hashtags affect Likes")
else:
    print("Fail to Reject Null Hypothesis")
    #Bonus
top_posts = df.sort_values(by='Likes', ascending=False).head(3)
print(top_posts)
```
